```python
import pandas as pd
import plotly.express as px
```

```python
spotify= pd.read_csv('spotify.csv')
spotify.head()
```

```python
spotify= spotify.drop(columns= ["Unnamed: 0"])
spotify.head()
```

```python
fig= px.line(spotify,\
                x='endDate',\
                y='msPlayed',\
                title= 'Spotify Listening Time by User per Day')
fig.show()
```

```python
fig= px.line(spotify,\
                x='endDate',\
                y='msPlayed',\
                color= 'userID',\
                title= 'Spotify Listening Time by User per Day')
fig.show()
```

```python
fig2= px.area(spotify,\
x='endDate',\
y='msPlayed',\
color= 'userID',\
title= 'Spotify Listening Time by User per Day')
fig.show()
```

```python
fig = px.area(spotify,\
x='day',\
y='msPlayed',\
animation_frame= 'month',\
animation_group= 'endDate',\
color= 'userID',\
title= 'Spotify Listening Time by User per Day')

fig.show()
```

```python
fig = px.area(spotify,\
x='day',\
y='msPlayed',\
animation_frame= 'month',\
animation_group= 'endDate',\
color= 'userID',\
title= 'Spotify Listening Time by User per Day')

fig["layout"].pop("updatemenus")

fig.show()
```

```python
fig = px.area(spotify,\
x='day',\
y='msPlayed',\
animation_frame= 'month',\
animation_group= 'endDate',\
range_x = [1,31],\
range_y = [0,55000000],\
color= 'userID',\
title= 'Spotify Listening Time by User per Day')

fig["layout"].pop("updatemenus")

fig.show()
```

```python
fig = px.area(spotify,\
x='day',\
y='msPlayed',\
animation_frame= 'month',\
animation_group= 'endDate',\
range_x = [1,31],\
range_y = [0,55000000],\
color= 'userID',\
title= 'Spotify Listening Time by User per Day')

fig["layout"].pop("updatemenus")
fig.update_layout(sliders=[dict(bgcolor='green', bordercolor= 'black')])

fig.show()
```

```python
def update_chart(y_variable):
    plt.figure(figsize=(12,8))

sns.countplot(x = 'Product', hue = y_variable,\
              data = data, palette = 'viridis')

plt.title('Distribution by Product{}'.format(y_variable))
plt.xlabel('Product')
plt.ylabel('Count')
plt.legend(title = y_variable)
plt.show()

y_variable_dropdown = widgets.Dropdown(
    options=y_variable,
    value = 'Age',
    description = 'y Variable:'
)

output = widgets.interactive_output(update_chart, {'y_variable': y_variable_dropdown})

display(widgets.HBox([y_variable_dropdown]))
display(output)
```
